```python
from qdrant_client import QdrantClient
from qdrant_client.models import Distance, VectorParams

client = QdrantClient(url="http://localhost:6333")
```

```python
# Connect to Qdrant (adjust host and port as needed)
client = QdrantClient(host="localhost", port=6333)
```

```python


client.create_collection(
    collection_name="test_collection",
    vectors_config=VectorParams(size=4, distance=Distance.DOT),
)
```

```python
client.delete_collection(collection_name="test_collection")
```

```python
# Get all collections
collections = client.get_collections()

# Print collection names
print("All collections:")
for collection in collections.collections:
    print(f"- {collection.name}")
```

```python
def quick_look(collection_name, limit=20):
    """–ë—ã—Å—Ç—Ä—ã–π –ø—Ä–æ—Å–º–æ—Ç—Ä —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ –∫–æ–ª–ª–µ–∫—Ü–∏–∏"""
    client = QdrantClient("localhost", port=6333)
    
    points = client.scroll(
        collection_name=collection_name,
        limit=limit,
        with_payload=True,
        with_vectors=False
    )[0]
    
    print(f"–ö–æ–ª–ª–µ–∫—Ü–∏—è: {collection_name}")
    print(f"–ü–æ–∫–∞–∑–∞–Ω–æ –ø–µ—Ä–≤—ã—Ö {len(points)} —Ç–æ—á–µ–∫:\n")
    
    for i, point in enumerate(points, 1):
        print(f"{i}. ID: {point.id}")
        print(f"   Payload: {point.payload}")
        print()

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ
quick_look("ai_cases")
```

```python
# –ü–æ–ª—É—á–∏—Ç—å –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫
count_result = client.count(
    collection_name="ai_cases"
)
print(f"Total points: {count_result.count}")
```
